```python
# ---- DATA LOADING -----

dataset_size = 'medium'

def load(dataset: str):
    return pd.read_csv(os.path.join(data_path, f'thedeep.{dataset_size}.{dataset}.txt'), names=["id", "text", "class"], index_col='id')

def load_labels():
    return pd.read_csv(os.path.join(data_path, f'thedeep.labels.txt'), names=["id", "name"], index_col='id')
    
data_train_original = load('train')
data_validation_original = load('validation')
data_test_original = load('test')
data_labels = load_labels()

display(Markdown('## Labels'))
display(data_labels)

display(Markdown('## Training data'))
display(data_train_original)

display(Markdown('## Validation data'))
display(data_validation_original)

display(Markdown('## Test data'))
display(data_test_original)
```
